# Start using Beautiful Soup to Scrape

# Basic Tag Extraction

```python
# Extract the title of the webpage
title = soup.title.text
print(f"Webpage Title: {title}")

# Extract all paragraph (p) tags
paragraphs = soup.find_all('p')
for p in paragraphs:
    print(p.text)
```

# CSS Selectors

```python
# Extract information using CSS selectors
headline = soup.select_one('h1').text
print(f"Headline: {headline}")

# Extract all elements with class 'example-class'
examples = soup.select('.example-class')
for example in examples:
    print(example.text)
```

# Navigation and Traversal

```python
# Navigate the HTML structure
first_link = soup.a
print(f"First Link Text: {first_link.text}")

# Traverse to the parent, next sibling, and previous sibling
parent_of_first_link = first_link.parent
next_sibling_of_first_link = first_link.next_sibling
previous_sibling_of_first_link = first_link.previous_sibling
```

# Advanced Extraction - Regular Expressions

```python
import re

# Extract all links containing 'example' using a regular expression
example_links = soup.find_all('a', href=re.compile('example'))
for link in example_links:
    print(link['href'])
```

## Download an Image




```python
# Done locally, but here is the code
import requests # pip install requests
from bs4 import BeautifulSoup as bs # pip install beautifulsoup4

# Load the webpage content
url = "https://keithgalli.github.io/web-scraping/"
r = requests.get(url+"webpage.html")

# Convert to a beautiful soup object
webpage = bs(r.content)

images = webpage.select("div.row div.column img")
image_url = images[0]['src']
full_url = url + image_url

img_data = requests.get(full_url).content
with open('lake_como.jpg', 'wb') as handler:
    handler.write(img_data)
```

## Solve the mystery challenge!




```python
files = webpage.select("div.block a")
relative_files = [f['href'] for f in files]


url = "https://keithgalli.github.io/web-scraping/"
for f in relative_files:
  full_url = url + f
  page = requests.get(full_url)
  bs_page = bs(page.content)
  secret_word_element = bs_page.find("p", attrs={"id": "secret-word"})
  secret_word = secret_word_element.string
  print(secret_word)
```
