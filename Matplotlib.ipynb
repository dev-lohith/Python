```python
import matplotlib.pyplot as plt
import numpy as np
```

# Line Plot

```python
# Generate random data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Plotting
plt.plot(x, y)
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

# Scatter Plot

```python
# Generate random data
x = np.random.rand(50)
y = np.random.rand(50)

# Scatter plot
plt.scatter(x, y, color='red', marker='o')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

# Bar Chart

```python
# Generate random data
categories = ['Category A', 'Category B', 'Category C']
values = np.random.randint(1, 10, size=len(categories))

# Bar chart
plt.bar(categories, values, color='skyblue')
plt.title('Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

# Histogram

```python
# Generate a random dataset
data = np.random.randn(1000)

# Histogram
plt.hist(data, bins=30, color='green', alpha=0.7)
plt.title('Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

# Pie Chart

```python
# Generate random data
labels = ['Label 1', 'Label 2', 'Label 3']
sizes = np.random.rand(len(labels))

# Pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['gold', 'lightcoral', 'lightskyblue'])
plt.title('Pie Chart')
plt.show()
```

# Subplots

```python
# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Plot 1: Line Plot
axes[0, 0].plot(x, y)
axes[0, 0].set_title('Line Plot')

# Plot 2: Scatter Plot
axes[0, 1].scatter(x, y, color='red', marker='o')
axes[0, 1].set_title('Scatter Plot')

# Plot 3: Bar Chart
axes[1, 0].bar(categories, values, color='skyblue')
axes[1, 0].set_title('Bar Chart')

# Plot 4: Histogram
axes[1, 1].hist(data, bins=30, color='green', alpha=0.7)
axes[1, 1].set_title('Histogram')

# Adjust layout
plt.tight_layout()
plt.show()
```
